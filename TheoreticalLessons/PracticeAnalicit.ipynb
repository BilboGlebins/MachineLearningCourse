```python
def darts(r1=2, r2=10):
    if r2 < r1:
        return ValueError
    large_circle = plt.Circle((0,0), r2, color='b', alpha=.8)
    small_circle = plt.Circle((0,0), r1, color='r', alpha=.8)
    
    fig, ax = plt.subplots(figsize=(5,5))
    ax.set_xlim(-r2, r2+1)
    ax.set_ylim(-r2, r2+1)
    ax.add_artist(large_circle)
    ax.add_artist(small_circle)
    ax.set_title('Darts')
    result_prob = 1 - (r1/r2) ** 2
    print(f'Вероятность того, что дротик не попал в красный круг {result_prob}')
    plt.show()
darts(r1=2, r2=10)
```

```python
def generate_data(n_points=20):
    x = np.linspace(-5,5, n_points)
    y = 10 * x - 7
    
    x_train = x[0::2].reshape(-1,1)
    y_train = y[0::2] + np.random.randn(int(n_points/2)) * 10
    
    x_test = x[1::2].reshape(-1,1)
    y_test = y[1::2] + np.random.randn(int(n_points/2)) * 10
    
    print(f'Generated {len(x_train)} train samples and {len(x_test)} test samples')
    return x, y, x_train, y_train, x_test, y_test
```

```python
x, y, x_train, y_train, x_test, y_test = generate_data(100)
```

```python
plt.figure(figsize=(10,5))

plt.plot(x, y, label='real')
plt.scatter(x_train, y_train, label='train')
plt.scatter(x_test, y_test, label='test')

plt.legend(loc='best')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()

model.fit(x_train, y_train)

w = model.coef_
b = model.intercept_
```
