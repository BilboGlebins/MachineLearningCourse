```python
g = 9.8
h0 = 100
v0 = 5

t = np.arange(0,5,0.1)

h = h0 - v0 * t - g * t ** 2 / 2

h[h < 0] = 0

plt.plot(t,h, label='Линия')

plt.xlabel('Время, с.')
plt.ylabel('Высота, м.')
plt.title(f'Cвободное падение с высоты {h0}м. и начальной скорости {v0}м/с.')
plt.legend()
```

```python
g = 9.8
h0 = 100
v0 = 5

t = np.arange(0,5,0.1)

h = h0 - v0 * t - g * t ** 2 / 2

h[h < 0] = 0

x = np.arange(0, 4*np.pi, 0.1) 
plt.plot(x, np.sin(x), label='Вода')

plt.plot(t,h, label='Нога')

plt.xlabel('Время, с.')
plt.ylabel('Высота, м.')
plt.title('Окунаем ногу в воду')
plt.legend()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
x = x[:, np.newaxis]

sin = np.sin(x)
cos= np.cos(x)

stacked = np.hstack([sin, cos])

plt.plot(x, stacked)
```

```python
for i, linestyle in enumerate(['-', '--', '-.', ':']):
    plt.plot(x, np.sin(x + i / 8 * np.pi), linestyle, label = linestyle)
```

```python
for i, color in enumerate(['orange', 'k', '0.75', '#FFDD44', (.2, .4, .8), 'orchid']):
    plt.plot(x, np.sin(x + i / 12 * np.pi), color, label=str(color))
plt.legend()
```

```python
x = np.arange(0, 4*np.pi, 0.1)

fig, ax = plt.subplots(1, 2, figsize=(16,5), sharey=True)
ax1, ax2 = ax

ax1.plot(4*np.sin(x))
ax2.scatter(x, 2*np.cos(x))
ax1.set_title('sin')
ax2.set_title('cos')

fig.tight_layout()
```

```python
x = np.arange(0, 4*np.pi, 0.1)

fig, ax = plt.subplots(2, 2, figsize=(16,5), sharey=True, sharex=True)
ax1, ax2 = ax[:, 1]

ax1.plot(np.sin(x))
ax2.plot(np.cos(x))
ax1.set_title('sin')
ax2.set_title('cos')

fig.tight_layout()
```
