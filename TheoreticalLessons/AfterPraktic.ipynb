-  \begin{equation}precision() =\frac{TP}{TP+FP}\end{equation}
-  \begin{equation}recall() =  \frac{TP}{TP+FN}\end{equation}
- F- \begin{equation}Accuracy = \frac{P*R}{P+R}\end{equation}
- Accuracy (   ) \begin{equation}accuracy =  \frac{TP+TN}{TP+TN+FP+FN}\end{equation} \begin{equation}Accuracy = \frac{P}{N}\end{equation}

```python
data = [[True,True], [True,True], [False,False], [False, True], [False,False]]
indexes = [i for i in range(1,6)]
frame = pd.DataFrame(data=data, index=indexes, columns=['',''])
print(frame)
accuracy = sk.metrics.accuracy_score(frame[''],frame[''])
recall = sk.metrics.recall_score(frame[''],frame['']) 
precision= sk.metrics.precision_score(frame[''],frame[''])

print("priciesion: %0.3f" % sk.metrics.precision_score(frame[''],frame['']))
print(f'priciesion={precision*100}%')
print(f'recall={recall*100}%')
print(f'accuracy={accuracy*100}%')

# priciesion = 66% = [1,2]/[1,2]+[4] = 2/3
# recall = 100% = [1,2]/[1,2]+[] = 2/2
# accuracy = 80% = [1,2]+[3,4]/[1,2]+[3,4]+[4]+[]= (2+2)/(2+2+1) = 4/5
```

```python
# ROC Curve
data = [[True,0.9], [True,0.8], [False,0.2], [False, 0.6], [False,0.1]]
indexes = [i for i in range(1,6)]
frame = pd.DataFrame(data=data, index=indexes, columns=['','()'])
print(frame)
# priciesion = 66% = [1,2]/[1,2]+[4] = 2/3
# recall = 100% = [1,2]/[1,2]+[] = 2/2
# accuracy = 80% = [1,2]+[3,4]/[1,2]+[3,4]+[4]+[]= (2+2)/(2+2+1) = 4/5
```
