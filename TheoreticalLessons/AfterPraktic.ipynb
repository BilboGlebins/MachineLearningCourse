- Формула (mae) mean_absolute_error \begin{equation}\frac{1}{n}\sum_{i=1}^{n} |y^{t}-y^{P}|\end{equation}
- Формула (mse) mean square error  \begin{equation}\frac{1}{n}\sum_{i=1}^{n} (y^{t}-y^{P})^{2}\end{equation}
- Формула (rmse) mean square error \begin{equation}\sqrt[2]{\frac{1}{n}\sum_{i=1}^{n} (y^{t}-y^{P})^{2}}\end{equation}
- Кванти́льная ошибка \begin{equation}QL = \frac{2*1,5+3*1,5+ 5*0,5}{3} = 3,3(3)\end{equation}
- Средняя абсолютная ошибка в процентах измеряется в % и поскольку невязка то чем меньше тем лучше \begin{equation}MAPE = \frac{1}{n}\sum_{i=1}^{n} |\frac{y^{t}-y^{P}}{y^{t}}|\end{equation} 

rmse и mae в одних единицах измерения а mse нет так как там присутствует квадарт (см*см = см2)\
mae - терпелив к выбросам так как отсутствуют квадраты\
mae позволяет допустить ошибки при расчётах\

потеря/издержки\

```python
# невязка
data = [[20,18], [15,12], [14,19]]
indexes = [i for i in range(1,4)]
frame = pd.DataFrame(data=data, index=indexes, columns=['Реальные','Предсказанные'])
print(frame)
frame_mae= sk.metrics.mean_absolute_error(frame['Реальные'],frame['Предсказанные'])
frame_mse = sk.metrics.mean_squared_error(frame['Реальные'],frame['Предсказанные'])


print(f'mae: ={frame_mae}')
print(f'mse: ={frame_mse}')



# mean square error mse
```

# Задачи Классификации

Метрики:
- точность \begin{equation}precision(тончность) =\frac{TP}{TP+FP}\end{equation}
- полнота \begin{equation}recall(полнота) =  \frac{TP}{TP+FN}\end{equation}
- F-мера \begin{equation}Accuracy = \frac{P*R}{P+R}\end{equation}
- Accuracy (Доля правильных ответов алгоритма) \begin{equation}accuracy =  \frac{TP+TN}{TP+TN+FP+FN}\end{equation} \begin{equation}Accuracy = \frac{P}{N}\end{equation}
- Roc curve
- precision-recall curve


accuracy - не показывает в какую сторону была произошла ошибка.
Несбалансированные классы

```python
data = [[True,True], [True,True], [False,False], [False, True], [False,False]]
indexes = [i for i in range(1,6)]
frame = pd.DataFrame(data=data, index=indexes, columns=['Реальные','Предсказанные'])
print(frame)
accuracy = sk.metrics.accuracy_score(frame['Реальные'],frame['Предсказанные'])
recall = sk.metrics.recall_score(frame['Реальные'],frame['Предсказанные']) 
precision= sk.metrics.precision_score(frame['Реальные'],frame['Предсказанные'])

print("priciesion: %0.3f" % sk.metrics.precision_score(frame['Реальные'],frame['Предсказанные']))
print(f'priciesion={precision*100}%')
print(f'recall={recall*100}%')
print(f'accuracy={accuracy*100}%')

# priciesion = 66% = [1,2]/[1,2]+[4] = 2/3
# recall = 100% = [1,2]/[1,2]+[] = 2/2
# accuracy = 80% = [1,2]+[3,4]/[1,2]+[3,4]+[4]+[]= (2+2)/(2+2+1) = 4/5
```

```python
# ROC Curve
data = [[True,0.9], [True,0.8], [False,0.2], [False, 0.6], [False,0.1]]
indexes = [i for i in range(1,6)]
frame = pd.DataFrame(data=data, index=indexes, columns=['Реальные','Предсказанные(Вероятность)'])
print(frame)
# priciesion = 66% = [1,2]/[1,2]+[4] = 2/3
# recall = 100% = [1,2]/[1,2]+[] = 2/2
# accuracy = 80% = [1,2]+[3,4]/[1,2]+[3,4]+[4]+[]= (2+2)/(2+2+1) = 4/5
```

# Кластеризация

Основанная на плотности пространственная кластеризация для приложений с шумами (англ. Density-based spatial clustering of applications with noise, <b>DBSCAN</b>) — это алгоритм кластеризации данных, который предложили Маритин Эстер, Ганс-Петер Кригель, Ёрг Сандер и Сяовэй Су в 1996. Это алгоритм кластеризации, основанной на плотности — если дан набор точек в некотором пространстве, алгоритм группирует вместе точки, которые тесно расположены (точки со многими близкими соседями), помечая как выбросы точки, которые находятся одиноко в областях с малой плотностью (ближайшие соседи которых лежат далеко). DBSCAN является одним из наиболее часто используемых алгоритмов кластеризации, и наиболее часто упоминается в научной литературе.

- DBSCAN
- Кластеризация методом k-средних: \begin{equation}p = \sqrt{\sum_{i=1}^{n}(x_n - x_n^{центроид})^{2}}\end{equation}

- иерархическая кластеризация
