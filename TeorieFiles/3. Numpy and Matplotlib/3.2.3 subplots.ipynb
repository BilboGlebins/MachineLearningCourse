```python
import numpy as np
import matplotlib.pyplot as plt
```

# Figure и Axis
Чтобы продолжить дальше, необходимо разделять понятия Figure и Axes.
* Figure - это рисунок, на котором размещаются объекты Axes
* Axes - оси, это область рисунка с парой осей x и y, на которых размещаются данные для отображения

Для начала работы с несколькими графиками на одном рисунке необходимо ознакомиться с функцией `plt.subplots()`. 
В качестве параметров она принимает количество `строк` и `столбцов` с отдельными графиками. Второй параметр можно опустить, тогда количество столбцов принимается равным единице. 
Функция `plt.subplots` возвращает объект типа figure, на котором будут рисоваться графики и список, содержащий созданные объекты Axes

_Попробуйте другие конфигурации осей изменяя параметры функции `plt.subplots`_

```python
_= figure, axes = plt.subplots(2)
print(figure)
print(axes)
```

Теперь о построении графиков. Ранее мы использовали функции `plt.plot()` и `plt.scatter()` без указания осей, коль скоро ось была одна.
Посмотрите, к какому результату приведёт использование этих функции в той же манере

_И снова попробуйте другие конфигурации осей изменяя параметры функции `plt.subplots()`, например (2, 2)_

```python
x = np.arange(0, 4*np.pi, 0.1)

_= figure, axes = plt.subplots(2)
plt.plot(np.sin(x))
```

Такое поведение закономерно, потому, что такие функции используют последние используемые оси. В данном случае, график строится на осях, которые были созданы последними.
Чтобы построить график на определённых осях, нужно вызывать метод `plot` соответствующего объекта:

_Попробуйте самостоятельно разобраться с индексацией осей для 2+ столбцов_

```python
x = np.arange(0, 4*np.pi, 0.1)

fig, axes = plt.subplots(2, 1)

axes[0].plot(x, np.sin(x))
_ = axes[1].plot(x, np.cos(x))
```

Каждый график (Axes) подписывается отдельно функцией axes.set_title().

_Посмотреть, как ведут себя подписи при использовании классического `plt.title()`_

```python
x = np.arange(0, 4*np.pi, 0.1)

fig, axes = plt.subplots(2, 1)
ax1, ax2 = axes

ax1.plot(x, np.sin(x))
ax1.set_title('sin(x)')

ax2.plot(x, np.cos(x))
_ = ax2.set_title('cos(x)')
```

Чтобы всё было красиво, можно воспользоватьсья функцией `fig.tight_layout()`. Она разместит элементы на рисунке так, чтобы они не пересекались:

```python
x = np.arange(0, 4*np.pi, 0.1)

fig, axes = plt.subplots(2, 1)
ax1, ax2 = axes
ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))

ax1.set_title('sin(x)')
_ = ax2.set_title('cos(x)')

fig.tight_layout()
```

# Общие оси
Давайте рассмотрим следующий пример:

```python
x = np.arange(0, 4*np.pi, 0.1)

fig, axes = plt.subplots(1, 2, figsize=(16, 5))  # Размер фигуры задаётся в сантиметрах в формате ширина, высота
ax1, ax2 = axes
ax1.plot(x, np.sin(x))
ax1.set_title('sin(x)')
ax2.plot(x, 2*np.sin(x))
_ = ax2.set_title('2sin(x)')
```

При таком отображении непросто заметить, что данные синусоиды имеют разные амплитуды. Однако, оси можно шарить. Для этого в функцию `plt.subplots()` передаются параметры `sharex` и `sharey` соответственно.

_Попробуйте самостоятельно придумать пример, где наглядно демонстрируется что стоит шарить ось x_

```python
x = np.arange(0, 4*np.pi, 0.1)

fig, axes = plt.subplots(1, 2, figsize=(16, 5), sharey=True)  # Делимся осью y
ax1, ax2 = axes
ax1.plot(x, np.sin(x))
ax1.set_title('sin(x)')
ax2.plot(x, 2*np.sin(x))
_ = ax2.set_title('2sin(x)')
```
