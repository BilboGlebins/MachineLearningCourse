```python
x = np.random.normal(loc=0, scale=1, size=10000)
count, threshold, bash = plt.hist(x, bins=20)

for i in range(10):
    print(f'  [{threshold[i]:2.4}:{threshold[i+1]:2.4}]  {int(count[i])}  ')
```

```python
x = np.random.rand(10000)
count, threshold, bash = plt.hist(x, bins=20)

for i in range(10):
    print(f'  [{threshold[i]:2.4}:{threshold[i+1]:2.4}]  {int(count[i])}  ')
```

```python
g = 9.8
h0 = 100
v0 = 5

t = np.arange(0,5,0.1)

h = h0 - v0 * t - g * t ** 2 / 2

h[h < 0] = 0

plt.plot(t,h, label='')

plt.xlabel(', .')
plt.ylabel(', .')
plt.title(f'C    {h0}.    {v0}/.')
plt.legend()
```

```python
g = 9.8
h0 = 100
v0 = 5

t = np.arange(0,5,0.1)

h = h0 - v0 * t - g * t ** 2 / 2

h[h < 0] = 0

x = np.arange(0, 4*np.pi, 0.1) 
plt.plot(x, np.sin(x), label='')

plt.plot(t,h, label='')

plt.xlabel(', .')
plt.ylabel(', .')
plt.title('   ')
plt.legend()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
x = x[:, np.newaxis]

sin = np.sin(x)
cos= np.cos(x)

stacked = np.hstack([sin, cos])

plt.plot(x, stacked)
```

```python
plt.scatter(x, sin)
plt.plot(x, cos, c='red')
```

```python
for i, linestyle in enumerate(['-', '--', '-.', ':']):
    plt.plot(x, np.sin(x + i / 8 * np.pi), linestyle = linestyle, label = linestyle)
```
